#### Análisis Exploratorio de Datos (EDA) del Dataset Iris

import libreries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar el dataset iris
iris = pd.read_csv('iris.csv')
```

```python
# Ver las primeras filas para entender la estructura
iris.head()
```

#### Part 1: Visualizaciones

1. Distribución del largo del pétalo (Histograma o KDE)

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=iris, x='petal_length', kde=True, bins=20)
plt.title('Distribución del Largo del Pétalo')
plt.xlabel('Largo del pétalo (cm)')
plt.ylabel('Frecuencia')
plt.show()
```

 El histograma muestra que la distribución del largo del pétalo es bimodal, lo que sugiere que hay dos grupos principales en los datos.

#### 2. Comparación del largo del sépalo por especie (Boxplot)

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=iris, x='species', y='sepal_length')
plt.title('Comparación del Largo del Sépalo por Especie')
plt.xlabel('Especie')
plt.ylabel('Largo del sépalo (cm)')
plt.show()
```

El boxplot revela claras diferencias en el largo del sépalo entre las especies. Iris setosa tiene los sépalos más cortos, mientras que virginica tiende a tener los más largos, con versicolor en un punto intermedio.

#### 3. Relación entre largo y ancho del pétalo (Scatter plot)

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species')
plt.title('Relación entre Largo y Ancho del Pétalo')
plt.xlabel('Largo del pétalo (cm)')
plt.ylabel('Ancho del pétalo (cm)')
plt.legend(title='Especie')
plt.show()
```

El scatter plot muestra una fuerte correlación positiva entre el largo y ancho del pétalo. Además, se observa que las diferentes especies forman grupos distintos en este espacio bidimensional.

#### 4. Frecuencia de cada especie (Bar chart)

```python
plt.figure(figsize=(8, 5))
iris['species'].value_counts().plot(kind='bar')
plt.title('Frecuencia de Cada Especie')
plt.xlabel('Especie')
plt.ylabel('Cantidad')
plt.xticks(rotation=0)
plt.show()
```

El gráfico de barras muestra que hay exactamente 50 muestras de cada especie en el dataset, lo que indica que está perfectamente balanceado en términos de distribución de clases.

#### 5. Correlación entre todas las variables numéricas (Heatmap)

```python
plt.figure(figsize=(8, 5))
sns.heatmap(iris.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title('Mapa de Calor de Correlaciones')
plt.show()
```

El heatmap revela que el largo y ancho del pétalo tienen una alta correlación positiva (0.96), mientras que la correlación entre el ancho del sépalo y el largo del pétalo es mucho menor (0.42).
